### Проверка полиномов f и g на биективность/транзитивность.
$ f(x) = 11 - 3x + 2x^2 $

$ g(x) = 13x - 1 $

```python
def check_bijective(z, fun):
    # биективность
    z_ring = set(range(2**2))
    print(f'Кольцо вычетов по модулю 2^2: {z_ring}')

    x = 0
    x_first = 0
    f_im = [set() for _ in range(2**2)]
    ans = set()
    
    while (x < 2**2 - 1):
        
        # x_first хранит начальное значение x
        # а сам x := f(x) на каждой итерации
        x_first = x
        while True:
            f_x = fun(x, z) % 4
            print(f'x = {x}; f(x) = {f_x}')
            # проверка окончания пройденного цикла
            if f_x in f_im[x_first]: 
                print(f'цикл от x = {x_first} кончился')
                break
            f_im[x_first].add(f_x)
            print(f'Цикл от {x_first}: {f_im[x_first]}')
            x = f_x
        
        # если найденный цикл не содержит вершины других циклов (то есть это вовсе не цикл),
        # и не является единственным, то
        # добавляем получившийся цикл в множество всех циклов
        if f_im[x].issubset(ans): 
            if f_im[x] == z_ring: print("Полином является одноцикловой перестановкой")                
            else: print("Полином не является перестановкой")
            break
        else: ans.update(f_im[x])
        # ищем новый x, который ещё не рассматривался
        if z_ring != ans: x = z_ring.difference(ans).pop()
        # иначе все элементы кольца вычетов рассмотрены
        else: break

    print('Объединение всех циклов: ', end='')
    [print(f'{i}:{f_im[i]}', end=' ') for i in range(2**2)]
    flag = True if ans == z_ring else False
    print(f'\nПолином образует кольцо вычетов {z_ring}' if flag else f'\nПолином не образует кольцо вычетов {z_ring}', end=' ')
    print('==> полином биективен\n' if flag else '==> полином не биективен')
```

```python
def check_transitive(z, fun):
    # транзитивность
    z_ring = set(range(2**3))
    print(f'Кольцо вычетов по модулю 2^3: {z_ring}')

    x = 0
    f_im = set()
    while x not in f_im:
        f_im.add(x)
        print(f'x = {x}; f(x) = {fun(x, z) % 8}')
        x = fun(x, z) % 8
    flag = True if f_im == z_ring else False
    print('Полином ' + ('' if flag else 'не ') + f'образует единственный цикл {z_ring}', end=' ')
    print('==> транзитивен\n' if flag else '==> не транзитивен')
```

Ввод коэффициентов в порядке убывания порядка степеней (e.g "2 -3 11" для первого полинома) через пробел ячейкой ниже:

```python
z = input().split(' ')
z = list(map(float, z))

polynom_str = ''
for i in range(len(z) - 1):
    polynom_str += f'{z[i]} * x^{len(z) - 1 - i} + '
polynom_str += f'{z[len(z) - 1]}'

f = lambda x, z : int(sum([z[i]*x**(len(z) - 1 - i) for i in range(len(z))]))
```

```python
print(f"Полином {polynom_str}:\n")

check_bijective(z, f)
check_transitive(z, f)
```

```python
z = input().split(' ')
z = list(map(float, z))

polynom_str = ''
for i in range(len(z) - 1):
    polynom_str += f'{z[i]} * x^{len(z) - 1 - i} + '
polynom_str += f'{z[len(z) - 1]}'

f = lambda x, z : int(sum([z[i]*x**(len(z) - 1 - i) for i in range(len(z))]))
```

### Проверка функции на транзитивность по модулю 256
$ f(x) = (x \oplus 1) \oplus (2 (x \land (1 + 2x) \land (3 + 4x) \land (7 + 8x) \land (15 + 16x) \land (31 + 32x) \land (63 + 64x))) \oplus (4(x^2+1)) $

```python
def check(fun):

    z_ring = list(range(2**8))
    print(f'Кольцо вычетов 2^8: {z_ring[:4]} ... {z_ring[-4:]}')
    z_ring = set(z_ring)

    x = 0
    f_im = set()
    while x not in f_im:
        f_im.add(x)
        print(f'x = {x}; f(x) = {fun(x) % 2**8}')
        x = fun(x) % 2**8
    print(f'==> Функция транзитивна по модулю 256' if f_im == z_ring else f'==> Функция не транзитивна по модулю 256')
```

```python
f = lambda x : (x ^ 1) ^ (2*(x & (1 + 2*x) & (3 + 4*x) & (7 + 8*x) & (15 + 16*x) & (31 + 32*x) & (63 + 64*x))) ^ (4*(x**2 + 1))
check(f)
```
