## Задание D, вариант 1

### Построить последовательность ван дер Корпута с помощью отображения Какутани-фон Неймана. Построить отображение на $Z_2$, индуцированное отображением Какутани-фон Неймана с помощью отображения Монна.

```python
import matplotlib.pyplot as plt
```

Последовательность ван дер Корпута $[0, 1]$:
$$0, \frac{1}{2}, \frac{1}{4}, \frac{3}{4}, \frac{1}{8}, \frac{5}{8}, \frac{3}{8}, \frac{7}{8}, \dots$$

Отображение Какутани - фон Неймана:
$$f(x) = x - 1 + \frac{1}{b^k} + \frac{1}{b^{k+1}}, \qquad 1 - \frac{1}{b^k} \leq x < 1 - \frac{1}{b^{k+1}}$$

```python
def kakutani_von_neumann_map(x):
    b = 2
    k = 0
    while not (x >= 1 - 1 / (b ** k) and x < 1 - 1 / (b ** (k + 1))):
        k += 1
    return x - 1 + 1 / (b ** k) + 1 / (b ** (k + 1))
```

```python
def monna_map(x):
    b = bin(x)[2:][::-1]
    ans = 0
    k = 0
    for i in b:
        ans += int(i) * (2** -(k + 1))
        k += 1
    return ans
```

Ввод n ячейкой ниже:

```python
n = int(input('n = '))
```

```python
x = []
y = []
tmp = 0
for _ in range(n):
    x.append(tmp)
    tmp = kakutani_von_neumann_map(tmp)
    y.append(tmp)
[print(f'x = {item_1}, T(x) = {item_2}') for item_1, item_2 in zip(x, y)]

plt.title(f'Van der Korput sequence\nvia Kakutani - von Neumann map, n = {n}')
plt.scatter(x, y)
plt.grid()
```

```python
x = []
y = []

for i in range(n + 1):
    x.append(i)
    y.append(monna_map(i))

plt.title(f'Van der Korput sequence\nvia Monna map, n = {n}')
plt.plot(x, y)
plt.scatter(x, y)
plt.grid()
```
